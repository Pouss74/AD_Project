```python
# Load the data specifying the column separator
data = pd.read_csv('DataCapstone.csv', delimiter=';', decimal=' ')
data.columns = data.columns.str.strip()

# Convert the data in the columns to float after replacing commas with periods and removing spaces
data['S&P 500 PRICE IN USD'] = data['S&P 500 PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)
data['GOLD PRICE IN USD'] = data['GOLD PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)
data['BITCOIN PRICE IN USD'] = data['BITCOIN PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)
data['ETHEREUM PRICE IN USD'] = data['ETHEREUM PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)

# Convert the 'Date' column to datetime type
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data.set_index('Date', inplace=True)

# Define the window for calculating the rolling standard deviation
window = 30  # it is common in finance

# List of assets to process
assets = {
    'S&P 500 PRICE IN USD': 'black',
    'GOLD PRICE IN USD': 'gold',
    'BITCOIN PRICE IN USD': 'blue',
    'ETHEREUM PRICE IN USD': 'green'
}

# Plot each asset's price evolution and volatility
fig, axs = plt.subplots(len(assets), 1, figsize=(12, 6 * len(assets)))
fig.suptitle('Price Evolution and Volatility for Each Asset', fontsize=16)

for ax, (asset, color) in zip(axs, assets.items()):
    # Create a secondary axis for the volatility
    ax2 = ax.twinx()

    # Plot price evolution on primary y-axis
    ax.plot(data.index, data[asset], color=color, label=f'{asset} Price')
    ax.set_ylabel('Price in USD')
    ax.set_title(f'{asset} Analysis')
    ax.legend(loc='upper left')

    # Calculate and plot volatility (rolling standard deviation of the prices) on secondary y-axis
    volatility = data[asset].rolling(window=window).std()
    ax2.plot(data.index, volatility, color='red', linestyle='--', label=f'{asset} Volatility (30-day rolling std)')
    ax2.set_ylabel('Volatility')
    ax2.legend(loc='upper right')

# Set common labels
plt.xlabel('Date')

# Improve layout
plt.tight_layout(rect=[0, 0.03, 1, 0.97])

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib.dates import date2num
import matplotlib.pyplot as plt

# Load the data specifying the column separator
data = pd.read_csv('DataCapstone.csv', delimiter=';', decimal=' ')

# Strip any extra spaces from the column names
data.columns = data.columns.str.strip()

# Convert the data in the columns to float after replacing commas with periods and removing spaces
data['S&P 500 PRICE IN USD'] = data['S&P 500 PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)
data['GOLD PRICE IN USD'] = data['GOLD PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)
data['BITCOIN PRICE IN USD'] = data['BITCOIN PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)
data['ETHEREUM PRICE IN USD'] = data['ETHEREUM PRICE IN USD'].str.replace(' ', '').str.replace(',', '.').astype(float)

# Convert the 'Date' column to datetime type
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Prepare data for regression
dates = date2num(data['Date']).reshape(-1, 1)

# Initialize the regression model
model = LinearRegression()

# List of assets to calculate regression and plot residuals
assets = ['S&P 500 PRICE IN USD', 'GOLD PRICE IN USD', 'BITCOIN PRICE IN USD', 'ETHEREUM PRICE IN USD']

# Prepare the plot for residuals
fig, axs = plt.subplots(4, 2, figsize=(14, 20), gridspec_kw={'width_ratios': [3, 1]})
fig.suptitle('Residual Analysis of Regression Models')

for idx, asset in enumerate(assets):
    y = data[asset].values.reshape(-1, 1)
    model.fit(dates, y)
    predictions = model.predict(dates)
    residuals = y - predictions

    # Scatter plot of residuals
    axs[idx, 0].scatter(data['Date'], residuals, color='blue', alpha=0.5, label='Residuals')
    axs[idx, 0].hlines(y=0, xmin=data['Date'].min(), xmax=data['Date'].max(), colors='red', linestyles='--', lw=2)
    axs[idx, 0].set_title(f'Residual Plot for {asset}')
    axs[idx, 0].set_xlabel('Date')
    axs[idx, 0].set_ylabel('Residuals')
    axs[idx, 0].legend()

    # Histogram of residuals
    axs[idx, 1].hist(residuals, bins=30, color='gray', alpha=0.7)
    axs[idx, 1].set_title(f'Residual Distribution for {asset}')
    axs[idx, 1].set_xlabel('Residuals')
    axs[idx, 1].set_ylabel('Frequency')

# Improve layout
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Function to test stationarity
def test_stationarity(timeseries):
    result = adfuller(timeseries.dropna())
    return result[1]  # p-value

# Applying transformation and differencing
def make_stationary(data, column):
    # Log transform
    data[f'log_{column}'] = np.log(data[column])
    
    # First differencing the log-transformed data
    data[f'log_diff_{column}'] = data[f'log_{column}'].diff()
    
    # Test stationarity
    p_value = test_stationarity(data[f'log_diff_{column}'])
    if p_value < 0.05:
        print(f"{column} is stationary after log transformation and first differencing with p-value = {p_value}")
    else:
        print(f"{column} is still not stationary, consider further differencing or transformations.")

# Applying transformations to each asset
for asset in assets:
    make_stationary(data, asset)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load data from CSV
data = pd.read_csv('DataCapstone.csv', delimiter=';', decimal=' ')

# Strip any extra spaces from the column names
data.columns = data.columns.str.strip()

# Convert the data in the columns to float after replacing commas with periods and removing spaces
columns_to_convert = ['S&P 500 PRICE IN USD', 'GOLD PRICE IN USD', 'BITCOIN PRICE IN USD', 'ETHEREUM PRICE IN USD']
for col in columns_to_convert:
    data[col] = data[col].str.replace(' ', '').str.replace(',', '.').astype(float)

# Convert the 'Date' column to datetime type
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Plot time series for each asset
def plot_time_series(data):
    for col in data.columns:
        if col in columns_to_convert:
            plt.figure(figsize=(15, 5))
            plt.subplot(1, 3, 1)
            plt.plot(data['Date'], data[col])
            plt.title(col + ' - Before Transformation')
            plt.xlabel('Date')
            plt.ylabel('Value')
            plt.grid(True)

            plt.subplot(1, 3, 2)
            plt.plot(data['Date'], np.log(data[col]))  # Apply np.log() to numeric data only
            plt.title(col + ' - Logarithmic Transformation')
            plt.xlabel('Date')
            plt.ylabel('Log(Value)')
            plt.grid(True)

            plt.subplot(1, 3, 3)
            plt.plot(data['Date'], np.log(data[col]).diff())
            plt.title(col + ' - Logarithmic Differencing Transformation')
            plt.xlabel('Date')
            plt.ylabel('Log Difference')
            plt.grid(True)

            plt.tight_layout()
            plt.show()

plot_time_series(data)

# Function to test stationarity
def test_stationarity(data):
    result = {}
    for col in data.columns:
        adf_result = adfuller(data[col])
        result[col] = {'ADF Statistic': adf_result[0], 'p-value': adf_result[1]}
    return pd.DataFrame(result)

# Check stationarity before transformation
stationarity_before = test_stationarity(data[columns_to_convert])
print("Stationarity Test Results (Before Transformation):\n", stationarity_before)

# Apply logarithm and differencing to make series stationary
data_log_diff = np.log(data[columns_to_convert]).diff().dropna()

# Check stationarity after transformation
stationarity_after = test_stationarity(data_log_diff)
print("Stationarity Test Results (After Logarithm and Differencing Transformation):\n", stationarity_after)

# Plot ACF and PACF for each asset before and after transformation
def plot_acf_pacf(data, title):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))
    for i, col in enumerate(data.columns):
        plot_acf(data[col], ax=axes[i//2, i%2], lags=20, title=f'ACF - {col}')
        plot_pacf(data[col], ax=axes[i//2, i%2], lags=20, title=f'PACF - {col}')
    plt.tight_layout()
    plt.suptitle(title, y=1.02)
    plt.show()

plot_acf_pacf(data[columns_to_convert], "ACF and PACF Before Transformation")
plot_acf_pacf(data_log_diff, "ACF and PACF After Transformation")
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('DataCapstone.csv', delimiter=';', decimal=',')
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data.set_index('Date', inplace=True)

# Specify that the data index should be treated as business days
data.index.freq = 'B'  # This sets the frequency to business days

# Convert all financial columns to float, ensuring non-numeric entries are handled
financial_columns = ['S&P 500 PRICE IN USD', 'GOLD PRICE IN USD', 'BITCOIN PRICE IN USD', 'ETHEREUM PRICE IN USD']
for column in financial_columns:
    data[column] = pd.to_numeric(data[column].astype(str).str.replace(',', '.').str.strip(), errors='coerce')

# Check for any remaining non-numeric entries and handle NaN values
data.dropna(inplace=True)  # Or you can use fillna() method depending on the requirement

# Function to fit ARIMA model and plot the results
def fit_arima(asset_data, order):
    model = ARIMA(asset_data, order=order)
    model_fit = model.fit()
    print(model_fit.summary())

    # Plot the actual vs fitted values
    plt.figure(figsize=(10, 4))
    plt.plot(asset_data, label='Actual')
    plt.plot(model_fit.fittedvalues, color='red', label='Fitted values')
    plt.title(f'ARIMA{order} - Fitted Model for {asset_data.name}')
    plt.legend()
    plt.show()

# Estimating ARIMA models for each asset
fit_arima(data['S&P 500 PRICE IN USD'], (1, 1, 1))
fit_arima(data['GOLD PRICE IN USD'], (0, 1, 1))
fit_arima(data['BITCOIN PRICE IN USD'], (1, 1, 1))
fit_arima(data['ETHEREUM PRICE IN USD'], (0, 1, 1))
```

```python
from pmdarima import auto_arima
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

# Load data
data = pd.read_csv('DataCapstone.csv', delimiter=';', decimal=',')
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data.set_index('Date', inplace=True)

# Convert all financial columns to float
financial_columns = ['S&P 500 PRICE IN USD', 'GOLD PRICE IN USD', 'BITCOIN PRICE IN USD', 'ETHEREUM PRICE IN USD']
for column in financial_columns:
    data[column] = pd.to_numeric(data[column].astype(str).str.replace(',', '.').str.strip(), errors='coerce')

data.dropna(inplace=True)

# Dictionary to store best models and parameters
best_models = {}

# Train model and evaluate it based on AIC, BIC, and RMSE
for column in financial_columns:
    print(f"Modeling {column}")
    
    # Split the data into train and test sets to calculate RMSE
    train, test = train_test_split(data[column], test_size=0.2, random_state=42, shuffle=False)

    # Auto ARIMA model
    model = auto_arima(train, seasonal=False, stepwise=True, suppress_warnings=True,
                       error_action="ignore", max_order=None, trace=False,
                       out_of_sample_size=len(test))  # Using test size for out-of-sample RMSE

    # Forecast and calculate RMSE
    predictions = model.predict(n_periods=len(test))
    rmse = sqrt(mean_squared_error(test, predictions))

    # Organize model results in dictionary
    best_models[column] = {
        'AIC': f"Params: {model.get_params()['order']}",
        'BIC': f"Params: {model.get_params()['order']}",
        'RMSE': f"Params: {model.get_params()['order']}, RMSE: {rmse:.2f}"
    }

    print(f"AIC: {model.aic()}, BIC: {model.bic()}, RMSE: {rmse}, Params: {model.get_params()['order']}")
    print(model.summary())
    print("\n\n")

# Convert the results to DataFrame for better visualization
results_df = pd.DataFrame(best_models).T  # Transpose to match the requested format
print(results_df)
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
from pandas.tseries.offsets import BDay

# Load data
data = pd.read_csv('DataCapstone.csv', delimiter=';', decimal=',')
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data.set_index('Date', inplace=True)
data.index.freq = 'B'

# Convert financial columns to float
financial_columns = ['BITCOIN PRICE IN USD', 'ETHEREUM PRICE IN USD']
for column in financial_columns:
    data[column] = pd.to_numeric(data[column].astype(str).str.replace(',', '.').str.strip(), errors='coerce')

# Handle NaN values
data.dropna(inplace=True)

# Define a function to forecast 1000 business days
def forecast_arima(series, order):
    # Fit the ARIMA model
    model = ARIMA(series, order=order)
    model_fit = model.fit()

    # Forecast the next 1000 business days
    forecast = model_fit.get_forecast(steps=1000)

    # Get the forecast mean
    forecast_mean = forecast.predicted_mean

    # Clip negative values to zero (simple method to avoid negatives)
    forecast_mean_clipped = forecast_mean.clip(lower=0)

    return forecast_mean_clipped

# Forecasting
bitcoin_forecast = forecast_arima(data['BITCOIN PRICE IN USD'], (5,1,2))
ethereum_forecast = forecast_arima(data['ETHEREUM PRICE IN USD'], (2,1,2))

# Create a date range for the next 1000 business days
future_dates = pd.date_range(data.index[-1] + BDay(), periods=1000, freq=BDay())

# Construct DataFrames for the forecasts
bitcoin_forecast_df = pd.DataFrame({'Date': future_dates, 'Bitcoin Forecast': bitcoin_forecast})
ethereum_forecast_df = pd.DataFrame({'Date': future_dates, 'Ethereum Forecast': ethereum_forecast})

# Output forecasts
print(bitcoin_forecast_df.head())
print(ethereum_forecast_df.head())
```
